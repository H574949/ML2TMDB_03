```python
newTrain = train[['budget', 'runtime', 'genres', 'original_language', 'popularity', 'revenue']]
```

```python
data = setup(data=newTrain, target='revenue', train_size=.99, numeric_imputation='median', normalize=True, normalize_method='robust')
```

```python
best_models = compare_models(n_select=4, sort='RMSE', include=['rf', 'gbr', 'en', 'huber', 'ridge'])
```

```python
## finetune model
tuned_best_models = [tune_model(i, optimize='RMSE', n_iter=2)for i in best_models]
```

```python
blender = stack_models(estimator_list=tuned_best_models[1:], meta_model=tuned_best_models[0])
```

```python

plot_model(blender)
```

```python
plot_model(blender, plot = 'error')
```

```python
final_model = finalize_model(blender)
```

```python
save_model(final_model,'box_office')
```
